This notebook plots some cosmological quantities that justify why LBG clustering and cross-correlation with CMB lensing does not depend much on $\Delta z$. I.e., because the weak lensing kernel, the linear growth factor, and the other geometric factors are relatively flat at high redshift, which is because $d\chi/dz$ and $dt/dz$ are both small at high redshifts.

```python
import pyccl as ccl
import numpy as np
from astropy.constants import c
import astropy.units as u
import matplotlib.pyplot as plt
```

```python
cosmo = ccl.Cosmology(Omega_c=0.25, Omega_b=0.05, h=0.67, sigma8=0.8, n_s=0.965)
c_over_H0 = (c / (cosmo["H0"] * u.km / u.s / u.Mpc)).to(u.Mpc)

z = np.linspace(1e-2, 6, 1000)
a = 1 / (1 + z)

D = ccl.growth_factor(cosmo, a)
chi = ccl.comoving_radial_distance(cosmo, a) * u.Mpc
dchi_dz = c_over_H0 / np.sqrt(cosmo["Omega_m"] * z**3 + (1 - cosmo["Omega_m"]))
Pi = D**2 * dchi_dz / chi * 1 / chi
dPi_dz = np.gradient(Pi, z)

t = cosmo.age_of_universe(a)
dt_dz = np.gradient(t, z)
dD_dt = np.gradient(D, t)
```

```python
fig = plt.figure(figsize=(7.5, 4), dpi=150)
gs = fig.add_gridspec(2, 6, hspace=0.4, wspace=1.5)

ax1 = fig.add_subplot(gs[0, 0:2])
ax1.plot(z, D**2)
ax1.set(xlabel="$z$", ylabel="$D^2$", ylim=(0, 1))

ax2 = fig.add_subplot(gs[0, 2:4])
ax2.plot(z, dchi_dz / chi)
ax2.set(xlabel="$z$", ylabel=r"$\chi^{-1} \, d\chi/dz$", ylim=(0, 5))

ax3 = fig.add_subplot(gs[0, 4:6])
ax3.plot(z, 1 / chi)
ax3.set(xlabel="$z$", ylabel=r"$\chi^{-1}$", ylim=(0, 1e-3))
ax3.ticklabel_format(axis="y", style="sci", scilimits=(0, 0))

ax4 = fig.add_subplot(gs[1, 1:3])
ax4.plot(z, Pi)
ax4.set(xlabel="$z$", ylabel="$\Pi$", yscale="log", ylim=(1e-7, 1e0))

ax5 = fig.add_subplot(gs[1, 3:5])
ax5.plot(z, dPi_dz / Pi)
ax5.set(xlabel="$z$", ylabel=r"$\Pi^{-1} \,d\Pi/dz$", ylim=(-10, 0))

axes = [ax1, ax2, ax3, ax4, ax5]
for ax in axes:
    ax.set(xticks=np.arange(0, 6, 1), xlim=(0, 6))
    ax.grid(axis="x", linestyle="--", alpha=0.5)

plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(
    1, 3, figsize=(7.5, 2), constrained_layout=True, dpi=120
)

ax1.plot(z, dD_dt)
ax1.set(xlabel="$z$", ylabel=r"$dD/dt$")

ax2.plot(z, dt_dz)
ax2.set(xlabel="$z$", ylabel=r"$dt/dz$")

ax3.plot(z, dD_dt * dt_dz)
ax3.set(xlabel="$z$", ylabel=r"$dD/dz$")

axes = [ax1, ax2, ax3]
for ax in axes:
    ax.set(xticks=np.arange(0, 6, 1), xlim=(0, 6))
    ax.grid(axis="x", linestyle="--", alpha=0.5)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(5, 2), constrained_layout=True, dpi=200)

ax1.plot(z, dchi_dz / 1e3)
ax1.set(xlabel="$z$", ylabel=r"$d\chi/dz$  [Gpc]", ylim=(0, 6))

ax2.plot(z, np.gradient(cosmo.lookback_time(a), z))
ax2.set(xlabel="$z$", ylabel=r"$dt/dz$  [Gyr]", ylim=(0, 15))

axes = [ax1, ax2]
for ax in axes:
    ax.set(xticks=np.arange(0, 6, 1), xlim=(0, 6))
    ax.grid(axis="x", linestyle="--", alpha=0.5)
```
